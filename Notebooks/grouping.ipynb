```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
static_df = pd.read_csv('../VED_Static_Data_ICE_HEV.csv')
ICEs = static_df[static_df['Vehicle Type'] == 'ICE']
weight = sorted(ICEs['Generalized_Weight'])

print(weight)
plt.hist(weight)
```

```python
configs_and_displacements = ICEs['Engine Configuration & Displacement']
displacements = configs_and_displacements.str.extract(r'(\d\.\d)')
print(displacements[0])
displacements = displacements.apply(float, axis=1)
displacements = sorted(displacements)
print(displacements)
plt.hist(displacements)
```
